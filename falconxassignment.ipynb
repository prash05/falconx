```python
import pandas as pd
import requests
import json
```

```python
query = """ query trades{
  swaps(
    orderBy: timestamp, orderDirection: desc) {
    pair{
      token0{
        symbol
      }
      token1{
        symbol
      }
      createdAtTimestamp
      id
      volumeToken0
      volumeToken1
      volumeUSD
      reserveUSD
      
    }
  
  }
}
"""
url = 'https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2'
r = requests.post(url, json={'query': query})
print(r.status_code)
print(r.text)
```

```python
json_data =json.loads(r.text)
```

```python
df_data =json_data['data']['swaps']
columns = ['Timestamp', 'liquidity', 'base_quant','quote_quant', 'volumeUSD','id'] 
#df = pd.DataFrame(df_data, columns=columns)
timestamp, liquidity, base_quant, quote_quant, volUSD, idn=[],[],[],[],[],[]
for x in df_data:
    if (x["pair"]["token0"]["symbol"]=="UNI"or x["pair"]["token0"]["symbol"]=="WETH")and(x["pair"]["token1"]["symbol"]=="UNI"or x["pair"]["token1"]["symbol"]=="WETH"):
        timestamp.append(x["pair"]["createdAtTimestamp"])
        liquidity.append(round(float(x["pair"]["reserveUSD"]),2))
        base_quant.append(round(float(x["pair"]["volumeToken0"]),2))
        quote_quant.append(round(float(x["pair"]["volumeToken1"]),2))
        volUSD.append(round(float(x["pair"]["volumeUSD"]),2))
        idn.append(x["pair"]["id"])
df= pd.DataFrame(list(zip(timestamp, liquidity, base_quant, quote_quant, volUSD, idn)),columns=columns)
df.head()
```
